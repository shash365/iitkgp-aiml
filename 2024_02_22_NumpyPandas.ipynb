# coming back to iris dataset

```python
sepal_length =  np.array(iris_df['sepal_length'])
sepal_width = np.array(iris_df['sepal_width'])
```

```python
mean_sepal_length = np.mean(sepal_length)
std_sepal_width = np.std(sepal_width)
print('Mean Sepal length: ', mean_sepal_length)
print('Standard deviation sepal width: ', std_sepal_width)
```

```python
large_sepals = sepal_length > mean_sepal_length
print(large_sepals)
print('Flowers with large sepals: ', sepal_length[large_sepals])
```

# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length VS Sepal Width')
plt.show()
```

```python
# correlation matrix
```

```python
# have to convert label to numeric and then we can use correlation matrix for heatmap

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
iris_df_encoded = iris_df.copy()
```

```python
iris_df_encoded.head(3)
```

```python
iris_df_encoded = iris_df_encoded[['sepal_length','sepal_width','petal_length','petal_width','area','species']]
```

```python
iris_df_encoded.head(3)
```

```python
iris_df_encoded['species'] =le.fit_transform(iris_df_encoded['species'])
```

```python
iris_df_encoded.head(3)
```

```python
iris_df_encoded.species.unique()
```

```python
correlation_matrix = iris_df_encoded.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation map of IRIS features and Label')
```
