```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetch data
data = pd.read_csv('LR_data.csv')
data.head()
```

```python
data.shape
```

```python
# Selecting features by analyzing the data. Here, in order to predict CO2 emission, car and model name are irrelevant.
# So Volumne and Wight will matter
```

```python
X = data[['Volume','Weight']]
y = data['CO2']
```

```python
# Import Lienar model

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

```python
# Lets predict CO2 emission when Volume and Weight car given for a car

co2_pred = lr.predict([[2500,1800]])
print(f'CO2 emission for a car when Volume :2500 and Weight: 1800 - {co2_pred}')
```

```python
# Lets take input from user and predict the CO2 emission for a car

Vol = int(input('Enter Volume of Car'))
Weg = int(input('Enter Weight of Car'))
co2_pred = lr.predict([[Vol,Weg]])
print(f'CO2 emission for the car when Volume :{Vol} and Weight: {Weg} - {co2_pred}')
```

Finally, we predicted that a car with 2.2-liter engine, and a weight of 1200 kg, will release approximately 105.93 grams of CO2 for per kilometer it drives.
